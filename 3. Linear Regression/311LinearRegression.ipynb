### Linear Regression
* Predict a numerical value.

Assumptions:
* Relationship between features x and target y is approzimately linear
* Noise is well behaved, following gausian distribution.

Model:
* How many features can be transformed into and estimate of the target.
* The assumption of linearity means that the expected value of the target(price) can be expressed as a weighted sum of features <br>
`y = w1x1 + ........ + wdxd + b` <br>
`y = w^T . x + b` <br>
* The goal of LR is to find the wight vector **w** and the bias term *b* such that, given features of a new data example sampled form the same distribution as **X**, <br>
the new example's label will be predicted with smallest error.
* Whatever instruments we use to observe the features **X** and labels **y**, there might be a small amount of measurement error.
* Two things 1.) measure the quality of model. 2.) a procedure for updating the model to improve its quality.

Loss function:
* Quantify the distance between real and predicted values of the target(non-negative number).
* Most common loss function is the squared error.

Minibatch Stochastic Gradient Descent
* Iteratively reducing the error by updating the parameters in the direction that incrementally lowers the loss function(which is avg of losses computed on every single example in the dataset), called *gradient descent*.
* The other extreme is to consider only a single example at a time and to take update steps based on one observation at a time (*Stochastic gradient descent*)
* SGD has drawbacks, both computationally and statistically.
* Processors are a lot faster multiplying and adding numbers than they are at moving data from main memory to processor cache.
* Takes a longer to process one sample at a time compared to full batch.
* A second problem is that some of the layers, such as batch normalization, only work well when we have access to more than one obs at a time
* Solution - Mini batch of observations(a number between 32 and 256).
* In each iteration t, we first randomly sample **Bt**, consisting of a fixed number **|B|**

### Vectorization of speed

* We want to process whole minibatches of examples simultaneously.
* Doing this efficiently requires that we vectorize the calculations and leverage fast linear algebra libraries rather than writing costly for-loops in Python

```python
%matplotlib inline
import math
import time
import numpy as np
import torch
from d2l import torch as d2l
```

```python
n=10000
a = torch.ones(n)
b = torch.ones(n) 
```

```python
## Using for-loop
c = torch.zeros(n)
t = time.time()
for i in range(n):
    c[i] = a[i] + b [i]
f' {time.time() - t:.5f} sec'
```

```python
## element wise operation
t = time.time()
d = a+b
f' {time.time() - t:.5f} sec'
```

Vectorized code often yields order of magnitude speedups
